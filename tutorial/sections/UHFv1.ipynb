Two electron integrals in Molecular Basis

```python
# Transformation from chemits (Gaussian) to physicists (McWeeny's book) notation of the two-electron integrals: rsgtu
## Spin convention:
# even orbitals are \alpha
# odd orbitals are \beta
rsgtu_ph = np.zeros((N_qubit,N_qubit,N_qubit,N_qubit))

for r in range(N_qubit):
    if (r & 1) == 0:
        a=-1
    else:
        a=1
    # r=0,..,N_qubit
    for s in range(N_qubit):
        if (s & 1) == 0:
            b=-1
        else:
            b=1
        for t in range(N_qubit):
            if (t & 1) == 0:
                c=-1
            else:
                c=1
            for u in range(N_qubit):
                if (u & 1) == 0:
                    d=-1
                else:
                    d=1
                    
                if (a*b>0 and c*d>0):
                    fact=1.
                else:
                    fact=0.
                
                rsgtu_ph[r][t][s][u] = fact * rsgtu[r][s][t][u].copy()
```

```python
# do the one body term as in McWeeny Eq. (9.2.12)
# H_core as in Szabo Eq. (3.153). Taken from the Gaussian output

H_core = np.zeros((N_AO+1,N_AO+1))
rFs    = np.zeros((N_qubit  ,N_qubit))

#   [AO][AO]
H_core[1][1]=-0.9962602
H_core[1][2]=-0.9386350  
H_core[1][3]=-0.8721280  
H_core[1][4]=-0.8102180
H_core[2][1]=-0.9386350  
H_core[2][2]=-0.9371630  
H_core[2][3]=-0.8102180  
H_core[2][4]=-0.8725020
H_core[3][1]=-0.8721280  
H_core[3][2]=-0.8102180  
H_core[3][3]=-0.9962602  
H_core[3][4]=-0.9386350
H_core[4][1]=-0.8102180  
H_core[4][2]=-0.8725020 
H_core[4][3]=-0.9386350  
H_core[4][4]=-0.9371630

H_core_MO = np.zeros((N_qubit,N_qubit))
for r in range(N_qubit):
    #r = 0,...,N_qubits-1
    for s in range(N_qubit):
        for mu in range(1,N_AO+1):
            for nu in range(1,N_AO+1):
                #print(r,s,mu,nu)
                H_core_MO[r][s] += coef_spinor[mu-1][r]*coef_spinor[nu-1][s] * H_core[mu][nu]

#print H_core_MO                
#print(np.round(H_core_MO,N_qubits))
#with spin information!

rFs = np.zeros((N_qubit,N_qubit))
H_core_MO_spin = np.zeros((N_qubit,N_qubit))

for r in range(N_qubit):
    if (r & 1) == 0:
        a=-1
    else:
        a=1
    for s in range(N_qubit):
        if (s & 1) == 0:
            b=-1
        else:
            b=1
            
        if a*b>0:
            fact=1.
        else:
            fact=0.
            
        rFs[r][s]            = fact * H_core_MO[r][s]
        H_core_MO_spin[r][s] = fact * H_core_MO[r][s]
        
        for i in range (N_occ):
            # i=1,...,N_occ
            #spin information already in rsgtu_ph
            rFs[r][s] += rsgtu_ph[r][i][s][i] - rsgtu_ph[r][i][i][s]

#print(np.round(rFs,N_qubit))
```

```python
import qutip as qu

################################################################################
# SORTING FUNCTION
# seq : the sorted array
# swap_counter : the number of swaps needed for sorting

def sort(seq):
    # print(seq)
    swap_counter = 0
    changed = True
    while changed:
        changed = False
        for i in range(len(seq) - 1):
            if seq[i] > seq[i + 1]:
                swap_counter += 1
                seq[i], seq[i + 1] = seq[i + 1], seq[i]
                changed = True
                # print(seq)
    return seq, swap_counter


################################################################################
# MAPPING FOR THE NORMAL ORDERING
#
# e.g. for 2 occupied orbitals
#
# OCCUPIED         | VIRTUAL
# a[0]    a[1]     | a[2]    a[3]    ...
# -1      -2       | 3       4
#                  |
# adag[0] adag[1]  | adag[2] adag[3] ...
# 1       2        | -3      -4
#
# In the case adag[0]*a[1] -> 1,-2 (sort)  -> (-1) -2,1     -> -a[1]*adag[0]
# it is right adag[0]*a[1] -> b[0]*bdag[1] -> -bdag[1]*b[0] -> -a[1]*adag[0]
#
################################################################################

def normal_order(N_qubits, N_occupied, array_to_be_ordered, array_mapping, printout = False, JW = True):

    if printout == True:

        print("Normal Ordering Starting")
        print("You are printing variables for steps of the normal ordering")
        print("Jordan - Wigner transformation is : ", JW)
################################################################################
## FUNCTION Arguments
#
# N_qubits : number of qubits
# N_occupied : number of occupied orbitals
# array_to_be_ordered : indices of the terms
# array_mapping : type of each operator (create or destroy operators)
# print : print the initial and the normal ordered terms (default: False)
#
# e.g. adag[0]*a[1]*adag[1]*a[3]
# array_to_be_ordered : [0,1,1,3]
# array_mapping : [adag, a, adag, a]
#
################################################################################

#   printout the starting term
    if printout:

        a_print = []
        adag_print = []

        for i in range(N_qubits):
            a_print.append("a"+str(i))
            adag_print.append("adag"+str(i))

        print_in = " "
        for i in range(len(array_to_be_ordered)):
            print_in += array_mapping[i]+str(array_to_be_ordered[i])+str(" ")
        print("Starting term :",print_in)

    #mapping to numbers for sorting
    a_enum = []
    adag_enum = []

    for i in range(N_qubits):

        if i < N_occupied:
            a_enum.append(-(i+1))
            adag_enum.append(i+1)

        else:
            a_enum.append(i+1)
            adag_enum.append(-(i+1))

    #print(a_enum)
    #print(adag_enum)

#   make the array to be sorted
    array_to_be_sorted = []

    for i in range(len(array_to_be_ordered)):

        if array_mapping[i] == "adag":
            array_to_be_sorted.append(adag_enum[array_to_be_ordered[i]])

        elif array_mapping[i] == "a":
            array_to_be_sorted.append(a_enum[array_to_be_ordered[i]])

# print(array_to_be_sorted)

#   sorting and counting the number of swaps
    sign = (-1)**sort(array_to_be_sorted)[1]
    array_sorted = sort(array_to_be_sorted)[0]

    # print(sign, array_sorted)

    #print the final term
    if printout == True:

        print_out = ""
        for i in array_sorted:
            if i in a_enum:
                print_out+= "a"+str(a_enum.index(i))+" "

            elif i in adag_enum:
                print_out += "adag"+str(adag_enum.index(i))+" "

        print("Final term : ", str(sign), print_out)

    #creating a and adag operators
    a_oper = []
    adag_oper = []

    for i in range(N_qubits):

        if i == 0:
            operator = qu.destroy(2)

        else:
            operator = qu.identity(2)

        for j in range(1,N_qubits):

            if j < i:
                operator = qu.tensor(operator,qu.identity(2))

            elif i == j:
                operator = qu.tensor(operator,qu.destroy(2))

            else:
                if JW == True:
                    operator = qu.tensor(operator,qu.sigmaz())
                else:
                    operator = qu.tensor(operator, qu.identity(2))

        a_oper.append(operator)
        adag_oper.append(operator.dag())

    #creating the ordered array operator
    sorted_operator = 1

    for i in array_sorted:
        if i in a_enum:
            sorted_operator *= a_oper[a_enum.index(i)]

        elif i in adag_enum:
            sorted_operator *= adag_oper[adag_enum.index(i)]

    sorted_operator = sign*sorted_operator

    return sorted_operator
```
